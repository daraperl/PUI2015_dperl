```python
census_files = []
establishments = {}
counter = 0
for i in years:
    #go to the zip file and open it as a csv 
    filepath = 'zbp%stotals.zip' % i
    filezip = zipfile.ZipFile(filepath) 
    years = pd.read_csv(filezip.open(filepath.replace('.zip','.txt')))
    #Change all of the column headers to upper case (some are already in upper case)
    years.columns = [x.upper() for x in years.columns]
    #extract only the zip code and the number of establishments
    years = years[['ZIP', 'EST']]
    #rename the column with the year number
    years = years.rename(columns = {'EST': fullyear[counter]})
    census_files.append(years)
    counter +=1

establishments[0] = census_files[0]

for i in range(1,20):
    establishments[i] = pd.merge(census_files[i], establishments[i-1], on=['ZIP']) 

census_df = establishments[19]
```

```python
#Array of NYC ZipCodes in Integer Form
nyczips = zipdata.postalCode.astype(int)
#print nyczips
```

```python
#filter for only NYC zipcodes in combined 

#This was something which I thought was necessary.  I now realize it is not helpful
#for this assignment
df = census_df.loc[census_df['ZIP'].isin(nyczips)]

df = df.reset_index(drop=True)
print df.shape
```

```python
#reset the index to the zipcode
df = df.set_index(['ZIP'])
#df
```

```python
#turn the data frame into an array
array = df.values
#reverse the data so it is going in chronological order
revarray = np.fliplr(array)
print revarray
```

```python
#remove NaNs
badindex = []
for i, est in enumerate(revarray):
    if np.isnan(est).any():
        badindex.append(i)  
clean_est = np.delete(revarray, badindex, 0)

print badindex
print clean_est
#check the array
d = revarray.shape
arrayfloat = revarray.astype(float)
print arrayfloat

#Note no NaNs were found
```

```python
#standardize (how similar are these shapes)

for i in range(d[0]):
    #get the row mean
    rowmean = arrayfloat[i].mean().astype(float)
    #get the row standard dev
    rowstd = arrayfloat[i].std().astype(float)
    #standardize each row
    for j in range(d[1]):
        arrayfloat[i][j] = ((arrayfloat[i][j] - rowmean)/rowstd).astype(float)
print arrayfloat.shape
```

```python
#Run a cluster Analysis using scipy kmeans with 2, 4, 6, and 8 clusters

cluster2 = scipy.cluster.vq.kmeans(arrayfloat, 2, iter=20, thresh=1e-05)
print "2 Clusters:" 
print cluster2
print ""


cluster4 = scipy.cluster.vq.kmeans(arrayfloat, 4, iter=20, thresh=1e-05)
print "4 Clusters:" 
print cluster4
print ""


cluster6 = scipy.cluster.vq.kmeans(arrayfloat, 6, iter=20, thresh=1e-05)
print "6 Clusters:" 
print cluster6
print ""

cluster8 = scipy.cluster.vq.kmeans(arrayfloat, 8, iter=20, thresh=1e-05)
print "8 Clusters:" 
print cluster8
print ""
```
